```python
import pandas as pd
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
data=pd.read_csv("resume/Resume/Resume.csv")
```

```python
resume_str=data["Resume_str"][0]
```

```python
word_tokenize(resume_str)
```

```python
a=0
def clean_data(text):
    text=str(text)
    text=re.sub('[^A-Za-z0-9]'," ",text)
    text=text.lower()
    text=nltk.word_tokenize(text)
    lm=WordNetLemmatizer()
    text=[lm.lemmatize(i) for i in text if not i in set(stopwords.words("english"))]
    text=" ".join(text)
    return text
```

```python
data["trasnformed_text"]=data["Resume_str"].map(clean_data)
```

```python
data.to_csv("resume_transformed.csv",index=False)
```
